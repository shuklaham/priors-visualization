```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.io import output_file
```

In this case, we are in the Jupyter notebook, so call `output_notebook()`. We only need to call this once, and all subsequent calls to `show()` will display inline in the notebook.

```python
output_notebook()
```

```python
date1 = ["Early October"]
brain_ct_yes1 = ["Diverticulosis","Lesions","IPMN","Lobe Consolidation","Pneumonia","Pleural effusion"]
brain_ct_no1 = ["Abscess","Dilatation"]
date2 = ["Mid September"]
brain_ct_yes2 = ["White matter","Microvascular ischemic","Parenchymal volume loss", "Frontal extra-axial spaces","Subdural hygromas"]
brain_ct_no2 = ["Acute infarct","Intracranial hemorrhage","Focal mass"]

date3 = ["Early September"]
brain_ct_yes3 = ["Cavernous malformation","Left frontal lesion"]
brain_ct_no3 = ["Intracranial hemorrhage","Ischemia"]
date4 = ["Mid August"]
brain_ct_yes4 = ["Resection cavities", "Internal Hemorrhagic products"]
brain_ct_no4 = ["Midline shift", "Downward herniation"]
```

```python
# create a new plot with default tools, using figure
p = figure(title="Visualization of Priors- Brain CTs",
           x_axis_label="Latest to Older >>>>>",
           y_axis_label="Absence/Presence",plot_width=950, plot_height=900)

# add a circle renderer with a size, color, and alpha

#Vertical lines
p.line([2,2], [-7,7], line_dash=[4, 4],line_width=1)
p.line([4,4], [-7,7], line_dash=[4, 4],line_width=1)
p.line([6,6], [-7,7], line_dash=[4, 4],line_width=1)
#p.line([2,2], [-6,6], line_dash=[4, 4], line_width=1)

#Horizontal line
p.line([0,8], [0,0], line_dash=[4, 4],line_width=1)

# x and y grids are off
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

#yes1 from Brain CT
p.circle([1, 1, 1, 1, 1, 1], [1, 2, 3, 4, 5, 6], size=[15,15,15,15,15,15], legend="Presence",line_color="white", fill_color="green", fill_alpha=0.5)

p.text([1, 1, 1, 1, 1, 1], [1.05+0.1, 2.05+0.1, 3.05+0.1, 4.05+0.1, 5.05+0.1, 6.05+0.1],
       text=brain_ct_yes1,text_color="#333333",
       text_align="center", text_font_size="10pt")

#yes2 from Brain CT
p.circle([3,3,3,3,3,3], [1, 2, 3, 4, 5], size=[15,15,15,15,15], line_color="white", fill_color="green", fill_alpha=0.5)

p.text([3,3,3,3,3,3], [1.05+0.1, 2.05+0.1, 3.05+0.1, 4.05+0.1,5.05+0.1],
       text=brain_ct_yes2,text_color="#333333",
       text_align="center", text_font_size="10pt")

#yes3 from Brain CT
p.circle([5, 5], [1, 2], size=[15,15], line_color="white", fill_color="green", fill_alpha=0.5)

p.text([5, 5], [1.05+0.1, 2.05+0.1],
       text=brain_ct_yes3,text_color="#333333",
       text_align="center", text_font_size="10pt")

#yes4 from Brain CT
p.circle([7,7], [1, 2], size=[15,15], line_color="white", fill_color="green", fill_alpha=0.5)

p.text([7,7], [1.05+0.1, 2.05+0.1],
       text=brain_ct_yes4,text_color="#333333",
       text_align="center", text_font_size="10pt")

#no1 from Brain CT
p.circle([1, 1], [-1, -2], size=[15,15], legend="Absence", line_color="white", fill_color="red", fill_alpha=1.0)

p.text([1, 1], [-1.05+0.2, -2.05+0.2],
       text=brain_ct_no1,text_color="#333333",
       text_align="center", text_font_size="10pt")

#no2 from Brain CT
p.circle([3, 3, 3, 3], [-1, -2, -3], size=[15,15,15,15,15], line_color="white", fill_color="red", fill_alpha=1.0)

p.text([3, 3, 3, 3], [-1.05+0.2, -2.05+0.2, -3.05+0.2],
       text=brain_ct_no2,text_color="#333333",
       text_align="center", text_font_size="10pt")

#no3 from Brain CT
p.circle([5, 5], [-1, -2], size=[15,15], line_color="white", fill_color="red", fill_alpha=1.0)

p.text([5, 5], [-1.05+0.2, -2.05+0.2],
       text=brain_ct_no3,text_color="#333333",
       text_align="center", text_font_size="10pt")

#no4 from Brain CT
p.circle([7, 7], [-1, -2], size=[15,15], line_color="white", fill_color="red", fill_alpha=1.0)

p.text([7, 7], [-1.05+0.2, -2.05+0.2],
       text=brain_ct_no4,text_color="#333333",
       text_align="center", text_font_size="10pt")

#Months of reports
p.text([1, 3, 5, 7], [7,7,7,7],
       text=[date1,date2,date3,date4],text_color="#333333",
       text_align="center", text_font_size="15pt")

#legend
p.legend.location = "top_right"

# for HTML output, uncomment below
output_file("past_brain_cts.html")
show(p) # show the results
```

All Bokeh markers accept `size` (measured in screen space units) as a property. Circles also have `radius` (measured in "data" space units).
